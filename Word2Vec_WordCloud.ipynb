```python
# -*- coding: utf-8 -*-

import logging
from gensim.models import word2vec
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from os import path
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def main():

    logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
    sentences = word2vec.LineSentence("../Word_Segmentation/Total_20180614_20170510_preprocessed_20180614_20170510.segmentated")
    model = word2vec.Word2Vec(sentences, size=50, sg=1)

    #保存模型，供日後使用
    model.save("word2vec.model")
    
    #Usage
    query_str = "馬刺"
    print("相似詞前 100 排序")
    res = model.most_similar(query_str,topn = 100)
    for item in res:
        print(item[0]+","+str(item[1]))

    #new_added = ['殘缺', 1]
    #res.append(new_added)
    #currdir = path.dirname(__file__)

    #text = open(path.join(d, 'constitution.txt')).read()
    #frequencies = [(u'健',100),('abc',4),('def',3),('gf',2),('rt',1)]
    #frequencies = dict(frequencies)
    frequencies = dict(res)
    mask = np.array(Image.open("/Users/coslate/Word2Vec/cloud.png"))

    # Generate a word cloud image 此处原为 text 方法，我们改用 frequencies 
    #wordcloud = WordCloud().generate(text)
    cloud = WordCloud(
        font_path='/Users/coslate/Downloads/NotoSerifCJKtc-hinted/NotoSerifCJKtc-Regular.otf',
        background_color='white',   # 背景颜色
        max_words=200,               # 最多显示词组个数
        max_font_size=200,           # 最大字号
        mask=mask,
        width=1980, height=1020
    )
    wordcloud = cloud.fit_words(frequencies)

    # Display the generated image:
    # the matplotlib way:
    plt.figure(figsize=(30,25))
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.tight_layout(pad=0)
    plt.show()
    wordcloud.to_file("/Users/coslate/Word2Vec/Generate_WC_Image/spurs.png")
        
        
if __name__ == "__main__":
    main()
```
