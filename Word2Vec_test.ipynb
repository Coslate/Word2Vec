```python
from gensim.models import word2vec

model = word2vec.Word2Vec.load("Generate_Model/Total_20180618_20170510_preprocessed_20180618_20170510.segmentated.sg_1.model")
```

```python
model.wv['勇士']
```

```python
query_str = "勇士"
print("相似詞前 100 排序")
res = model.most_similar(query_str,topn = 100)
for item in res:
    print(item[0]+","+str(item[1]))
```
